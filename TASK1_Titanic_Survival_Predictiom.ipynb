import the necessary libs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

data loading/processing

```python
titanic_data = pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
titanic_data.head()
```

```python
titanic_data.isnull().sum()
```

```python
#remove missing/null values
titanic_data = titanic_data.drop(columns='Cabin',axis=1)
```

```python
#replacing missing values with mean number
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)
```

```python
titanic_data.info()
```

```python
titanic_data.isnull().sum()
```

```python
#lets fix embarked
print(titanic_data['Embarked'].mode() )
```

```python
print(titanic_data['Embarked'].mode()[0] )
```

```python
#replace the mode value with the missing value
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0],inplace=True)
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.replace({'Sex':{'male':0, 'female' :1}, 'Embarked':{'S':0,'C':1,'Q':2}})
```

```python
X = titanic_data.drop(columns = ['PassengerId','Name','Ticket','Survived'], axis=1)
Y = titanic_data['Survived']
```

```python
print(X)
```

```python
print(Y)
```

Spliting data into test and train data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

```python
model = LogisticRegression()
```

```python
#use the train data on logisticregression model
model.fit(X_train, Y_train)
```

Evaluating and Testing model

```python
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
#check accuracy of test data
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data:', test_data_accuracy)
```
